## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error
```

## Data Collection and Processing

```python
# Load the dataset
gold_dataset = pd.read_csv("./Datasets/gldpricedata.csv")
gold_dataset.head()
```

```python
gold_dataset.tail()
```

```python
# Getting basic information about the data
gold_dataset.info()
```

```python
# Check missing values
gold_dataset.isna().sum()
```

```python
# Statistical measures of the data
gold_dataset.describe()
```

 ## Correlation

```python
correlation = gold_dataset.corr()
```

```python
# Constructing a Heatmap
sns.heatmap(data=correlation, annot=True, annot_kws={"size":8}, fmt=".1f", cmap="Blues")
plt.show()
```

```python
# Correlation values of GLD
print(correlation["GLD"])
```

```python
# Check the distribution of GLD price
sns.distplot(gold_dataset["GLD"], color="b")
```

## Split the Features and Target

```python
x = gold_dataset.drop(columns=["Date", "GLD"], axis=1)
y = gold_dataset["GLD"]
```

## Separate Test Train

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

## Model Training

### Random Forest Regressor

```python
model = RandomForestRegressor()
```

```python
# Training the model
model.fit(x_train, y_train)
```

## Model Evaluation

```python
y_pred = model.predict(x_test)
```

```python
# R squared error
error_score = r2_score(y_test, y_pred)
error_score
```

## Compare the Actual and predicted values in a plot

```python
# Plot a subset of data for better visualization
subset_size = 100  # Adjust the size as needed
y_test_subset = y_test[:subset_size].reset_index(drop=True)
y_pred_subset = y_pred[:subset_size]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y_test_subset, color="blue", label="Actual Values", marker='o', linestyle='dashed')
plt.plot(y_pred_subset, color="green", label="Predicted Values", marker='x', linestyle='dashed')
plt.title("Actual Price vs Predicted Price")
plt.xlabel("Number of values")
plt.ylabel("GLD Price")
plt.legend()
plt.show()
```

```python
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
# Root mean squared error
rmse = np.sqrt(mse)
rmse
```
